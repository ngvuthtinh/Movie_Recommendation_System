# Normalize Movie Data for Insertion into Database

### Import Libraries

```python
import pandas as pd
from typing import Dict
from pathlib import Path
```

### Read the Data From CSV

```python
data_path = Path('data/movie_data.csv')
df = pd.read_csv(data_path)
```

```python
df.head(50)
```

### Normalize Data

```python
columns_to_keep = [
    "title", "vote_average", "vote_count", "status",
    "release_date", "runtime", "adult", "backdrop_path", "original_language",
    "overview", "popularity", "poster_path", "tagline", "genres",
    "production_companies", "production_countries", "spoken_languages", "keywords"
]

cleaned_df = df[columns_to_keep].copy()
cleaned_df.insert(0, 'movie_id', range(1, len(cleaned_df) + 1))  # Insert 'id' column at the beginning
```

```python
cleaned_df.head(10)
```

```python
def extract_multivalues(df: pd.DataFrame, column_name: str) -> pd.DataFrame:
    """
    Extracts multi-value columns into separate rows.
    """
    processed_df = df[column_name].str.split(", ").explode().reset_index(drop=True)
    processed_df = processed_df.to_frame(name=column_name).drop_duplicates()

    processed_df.insert(0, 'id', range(1, len(processed_df) + 1))
    return processed_df

def create_associative(df: pd.DataFrame, processed_df: pd.DataFrame) -> pd.DataFrame:
    """
    Creates an associative table for multi-value columns by connecting movie_id with processed_df id.
    """
    exploded_df = df.copy()
    column_name = processed_df.columns[1]

    # Split and explode the values
    exploded_df[column_name] = exploded_df[column_name].str.split(', ')
    exploded_df = exploded_df.explode(column_name).reset_index(drop=True)

    # Merge with processed_df to get the corresponding ids
    associative_table = (
        exploded_df[['movie_id', column_name]]
        .merge(processed_df, how='left', left_on=column_name, right_on=column_name)
        [['movie_id', 'id']]
        .rename(columns={'id': f"{column_name}_id"})
    )

    return associative_table.drop_duplicates().reset_index(drop=True)
```

```python
# For testing
a = extract_multivalues(cleaned_df, "production_countries")
b = create_associative(cleaned_df, a)
```

```python
movie_columns = [
    "movie_id", "title", "vote_average", "vote_count", "status", "release_date",
    "runtime", "adult", "backdrop_path", "original_language",
    "overview", "popularity", "poster_path", "tagline"
]

columns_to_normalize = [
    "genres", "production_companies", "production_countries",
    "spoken_languages", "keywords"
]
```

```python
result = []

movie_df = cleaned_df[movie_columns].copy()
result.append({'movie': movie_df})

for column in columns_to_normalize:
    processed_df = extract_multivalues(cleaned_df, column)
    associative_table = create_associative(cleaned_df, processed_df)

    result.append({f"{column}_table": processed_df})
    result.append({f"movie_{column}": associative_table})
```

# Save the Result to Pickle File

```python
import pickle

output_path = Path('data/normalized_movie_data.pkl')

with open(output_path, 'wb') as f:
    pickle.dump(result, f)

print(f"Normalized data saved to {output_path}")
```
